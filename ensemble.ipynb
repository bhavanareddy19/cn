```python
import warnings, itertools, pandas as pd, numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection    import train_test_split
from sklearn.preprocessing      import (StandardScaler, OrdinalEncoder,
                                        OneHotEncoder)
from sklearn.compose            import ColumnTransformer
from sklearn.pipeline           import Pipeline
from sklearn.metrics            import (accuracy_score,
                                        confusion_matrix,
                                        ConfusionMatrixDisplay,
                                        classification_report)
from sklearn.ensemble           import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model       import LogisticRegression
from sklearn.ensemble           import VotingClassifier

warnings.filterwarnings("ignore")

DATA_PATH = Path("cleaned_data.csv")           
keep      = ["Applications Developer", "Software Developer",
             "Software Engineer",     "Web Developer"]

df = (pd.read_csv(DATA_PATH)
        .query("SuggestedJobRole in @keep")
        .reset_index(drop=True))

print("RAW DATA (first 5 rows)"); display(df.head())

num_cols = ["Logical quotient rating", "hackathons",
            "coding skills rating",    "public speaking points"]

ordinal_cols = ["reading and writing skills",
                "memory capability score"]
ordinal_map  = [["poor", "medium", "excellent"]] * len(ordinal_cols)

binary_cols  = ["Management or Technical",
                "self-learning capability?",     "Extra-courses did",
                "certifications",                "workshops",
                "Taken inputs from seniors or elders",
                "hard/smart worker",             "worked in teams ever?",
                "Introvert"]

target = "SuggestedJobRole"
X = df[num_cols + ordinal_cols + binary_cols]
y = df[target]

preprocess = ColumnTransformer([
    ("num",  StandardScaler(),                      num_cols),
    ("ord",  OrdinalEncoder(categories=ordinal_map), ordinal_cols),
    ("bin",  OneHotEncoder(drop="if_binary"),        binary_cols)
])

preprocess.fit(X)
X_prev = preprocess.transform(X.iloc[:5])
if hasattr(X_prev, "toarray"):     
    X_prev = X_prev.toarray()
print("\nAFTER TRANSFORMATION (first 5 rows, first 10 cols)")
display(pd.DataFrame(X_prev[:, :10]))

X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42)

print(f"\nTrain / Test sizes  →  {X_tr.shape[0]} train  |  {X_te.shape[0]} test")

# 1) Transform both splits
X_tr_t = preprocess.transform(X_tr)
X_te_t = preprocess.transform(X_te)

# 2) If the result is sparse, convert to dense
if hasattr(X_tr_t, "toarray"):
    X_tr_t = X_tr_t.toarray()
    X_te_t = X_te_t.toarray()

# 3) Grab feature names
feat_names = preprocess.get_feature_names_out()

# 4) Display first five rows of each
print(">>> TRANSFORMED TRAIN (first 5 rows)")
display(pd.DataFrame(X_tr_t[:5], columns=feat_names))

print("\n>>> TRANSFORMED TEST (first 5 rows)")
display(pd.DataFrame(X_te_t[:5], columns=feat_names))

rf  = RandomForestClassifier(n_estimators=200,        random_state=0)
ada = AdaBoostClassifier    (n_estimators=150,        random_state=0)
log = LogisticRegression    (max_iter=1000, multi_class="multinomial")
vote = VotingClassifier(
            estimators=[("rf", rf), ("ada", ada), ("lr", log)],
            voting="soft")

models = {
    "Random Forest": rf,
    "AdaBoost"     : ada,
    "Soft Voting"  : vote
}

results = []
for name, base_clf in models.items():
    clf = Pipeline([("prep", preprocess), ("clf", base_clf)])
    clf.fit(X_tr, y_tr)
    y_pred = clf.predict(X_te)
    acc    = accuracy_score(y_te, y_pred)
    results.append((name, acc))
    
    print(f"\n{name}  –  accuracy = {acc:.4f}")
    print(classification_report(y_te, y_pred, zero_division=0))
    
    cm = confusion_matrix(y_te, y_pred, labels=clf.classes_)
    ConfusionMatrixDisplay(cm, display_labels=clf.classes_)\
        .plot(cmap="Blues", xticks_rotation=45)
    plt.title(f"{name}  (acc={acc:.2%})")
    plt.tight_layout(); plt.show()

summary = pd.DataFrame(results, columns=["Model", "Accuracy"])\
            .sort_values("Accuracy", ascending=False)
print("\n=== Ensemble accuracy summary ===")
display(summary)
```
