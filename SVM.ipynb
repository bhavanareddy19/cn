```python
import itertools, warnings, pandas as pd, numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (StandardScaler, OrdinalEncoder,
                                   OneHotEncoder, LabelEncoder)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, classification_report,
                             confusion_matrix, ConfusionMatrixDisplay)

warnings.filterwarnings("ignore") 

DATA_PATH = Path("cleaned_data.csv")         
keep      = ["Applications Developer", "Software Developer",
             "Software Engineer",     "Web Developer"]

df = (pd.read_csv(DATA_PATH)
        .query("SuggestedJobRole in @keep")
        .reset_index(drop=True))

print("RAW DATA (first 5 rows)")
display(df.head())             

num_cols = ["Logical quotient rating", "hackathons",
            "coding skills rating",    "public speaking points"]

ordinal_cols = ["reading and writing skills",
                "memory capability score"]
ordinal_map  = [["poor", "medium", "excellent"]] * len(ordinal_cols)

binary_cols  = ["Management or Technical",
                "self-learning capability?",     "Extra-courses did",
                "certifications",                "workshops",
                "Taken inputs from seniors or elders",
                "hard/smart worker",             "worked in teams ever?",
                "Introvert"]

target = "SuggestedJobRole"
X = df[num_cols + ordinal_cols + binary_cols]
y = df[target]

preprocess = ColumnTransformer([
    ("num",  StandardScaler(),                      num_cols),
    ("ord",  OrdinalEncoder(categories=ordinal_map), ordinal_cols),
    ("bin",  OneHotEncoder(drop="if_binary"),        binary_cols)
])

preprocess.fit(X)
X_preview = preprocess.transform(X.iloc[:5])
if hasattr(X_preview, "toarray"):  # sparse?
    X_preview = X_preview.toarray()
print("\nAFTER TRANSFORMATION (first 5 rows, first 10 cols)")
display(pd.DataFrame(X_preview[:, :10]))

X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42)

print(f"\nTrain-test sizes →  {X_tr.shape[0]} train  |  {X_te.shape[0]} test")
print("\nClass counts – TRAIN");  display(y_tr.value_counts())
print("\nClass counts – TEST");   display(y_te.value_counts())

kernels = ["linear", "rbf", "poly"]
C_vals  = [0.1, 1, 10]
results = []

for ker, C in itertools.product(kernels, C_vals):
    clf = Pipeline([
        ("prep", preprocess),
        ("svm",  SVC(kernel=ker, C=C, gamma="scale",
                     class_weight="balanced"))
    ])
    clf.fit(X_tr, y_tr)
    y_pred = clf.predict(X_te)
    acc    = accuracy_score(y_te, y_pred)
    results.append((ker, C, acc))
    
    print(f"\n────────────────────────────────────────────────────────")
    print(f"SVM {ker:<6}  C={C}   –   accuracy = {acc:.4f}")
    print(classification_report(y_te, y_pred, zero_division=0))
    
    cm = confusion_matrix(y_te, y_pred, labels=clf.classes_)
    ConfusionMatrixDisplay(cm, display_labels=clf.classes_)\
        .plot(cmap="Blues", xticks_rotation=45)
    plt.title(f"SVM {ker}  C={C}")
    plt.tight_layout(); plt.show()

summary = (pd.DataFrame(results, columns=["Kernel", "C", "Accuracy"])
             .sort_values("Accuracy", ascending=False))
print("\n=== Accuracy summary ===")
display(summary)

best_per_kernel = (summary
                   .groupby("Kernel")
                   .first()
                   .reset_index())

pca2d = Pipeline([("prep", preprocess), ("pca", PCA(n_components=2))])
X_tr2d = pca2d.fit_transform(X_tr)
labels_encoded = LabelEncoder().fit_transform(y_tr)

def plot_decision_surface(kernel, C):
    svm2d = SVC(kernel=kernel, C=C, gamma="scale",
                class_weight="balanced")
    svm2d.fit(X_tr2d, y_tr)
    x_min, x_max = X_tr2d[:,0].min()-1, X_tr2d[:,0].max()+1
    y_min, y_max = X_tr2d[:,1].min()-1, X_tr2d[:,1].max()+1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 400),
                         np.linspace(y_min, y_max, 400))
    Z_lbl = svm2d.predict(np.c_[xx.ravel(), yy.ravel()])
    Z_int = LabelEncoder().fit(Z_lbl).transform(Z_lbl).reshape(xx.shape)

    plt.figure(figsize=(6,4))
    plt.contourf(xx, yy, Z_int, alpha=.25, cmap="Pastel2",
                 levels=len(np.unique(Z_int)))
    scatter = plt.scatter(X_tr2d[:,0], X_tr2d[:,1],
                          c=labels_encoded, cmap="tab10",
                          s=20, edgecolor="k")
    plt.title(f"Decision boundary – {kernel}  C={C}")
    plt.xlabel("PCA-1"); plt.ylabel("PCA-2")
    plt.tight_layout(); plt.show()

print("\nDecision-boundary visualisations:")
for _, row in best_per_kernel.iterrows():
    plot_decision_surface(row["Kernel"], row["C"])
```
