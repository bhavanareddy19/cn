<h3>Data is Imbalance so we have to balance it in Data preprocesing</h3>

```python
ax = data["Suggested Job Role"].value_counts().plot(kind = "bar", title = "Count of Job Role", figsize = (12,6))
ax.set_xlabel("Job Role")
plt.show()
```

```python
num_col = []
for col in numerical_features:
    if col.find("Acedamic percentage in ") != -1:
        num_col.append(col.replace("Acedamic percentage in ", ""))
    elif col.find("percentage in ") != -1:
        num_col.append(col.replace("percentage in ", ""))
    elif col.find("Percentage in ") != -1:
        num_col.append(col.replace("Percentage in ", ""))
    else:
        num_col.append(col)
print(num_col)
```

```python
#Creating a Numerical features dataset to find out different trends
numerical_data = pd.DataFrame(np.array(data[numerical_features]), columns = num_col)
```
